```python
from dash import Dash, html, dcc, callback, Output, Input
import plotly.express as px
import pandas as pd
```

```python
df_temp = pd.read_csv(
    '../../Data/data_fix_temp.txt')
df_wmoid = pd.read_excel(
    '../../Data/daftar_wmoid.xlsx')
df_wmoid = df_wmoid.rename(columns={'WMOID': 'lokasi'})
df_wmoid = df_wmoid[['lokasi', 'Nama UPT']]
```

```python
# Rapihkan data nya
df_temp_sorted = df_temp.copy()
df_temp_sorted['Date'] = pd.to_datetime(df_temp['Date'])
df_temp_sorted = df_temp.set_index('Date')
df_temp_sorted = df_temp_sorted.sort_index()
# df_temp_sorted = df_temp_sorted.sort_values(by='Date')
df_temp_sorted
```

```python
import geopandas
import json
import dash_leaflet.express as dlx
import dash_leaflet as dl


df_map = df_wmoid.merge(df_temp, on='lokasi')
df_map = df_map[['lokasi', 'Nama UPT', 'LON', 'LAT']].drop_duplicates()
# Make geopandas geometry for coordinates
geometry = geopandas.points_from_xy(df_map.LON, df_map.LAT)
upt_gpd = geopandas.GeoDataFrame(df_map, geometry=geometry)
upt_gpd = upt_gpd.reset_index(drop=True)

geojson = json.loads(upt_gpd.to_json())
geobuf = dlx.geojson_to_geobuf(geojson)
upt = dl.GeoJSON(
    data=geobuf,
    id='geojson',
    format='geobuf',
    cluster=True
)
```

```python
from xgboost import XGBRegressor
model_xgb = XGBRegressor()
```

```python
model_xgb.load_model('../Dash-Monas/xgb_tuned_08_noShuffle_2.json')
```

```python
ina_nwp_input_filtered = ina_nwp_input.drop(columns=['Date', 'LAT', 'LON', 'prec_nwp']) 
ina_nwp_input_filtered = ina_nwp_input_filtered.rename(
  columns={
    'suhu2m(degC)' : 'suhu2m.degC.',
    'dew2m(degC)' : 'dew2m.degC.',
    'rh2m(%)' : 'rh2m...',
    'wspeed(m/s)' : 'wspeed.m.s.',
    'wdir(deg)' : 'wdir.deg.',
    'lcloud(%)' : 'lcloud...',
    'mcloud(%)' : 'mcloud...' ,
    'hcloud(%)' : 'hcloud...',
    'surpre(Pa)' : 'surpre.Pa.' ,
    'clmix(kg/kg)' : 'clmix.kg.kg.' ,
    'wamix(kg/kg)' : 'wamix.kg.kg.' ,
    'outlr(W/m2)' : 'outlr.W.m2.' ,
    'pblh(m)' : 'pblh.m.',
    'lifcl(m)' : 'lifcl.m.' ,
    'cape(j/kg)' : 'cape.j.kg.' ,
    'mdbz' : 'mdbz' ,
    't950(degC)' : 't950.degC.' ,
    'rh950(%)' : 'rh950...',
    'ws950(m/s)' : 'ws950.m.s.' ,
    'wd950(deg)' : 'wd950.deg.' ,
    't800(degC)' : 't800.degC.' ,
    'rh800(%)' : 'rh800...' ,
    'ws800(m/s)' : 'ws800.m.s.',
    'wd800(deg)' : 'wd800.deg.' ,
    't500(degC)' : 't500.degC.' ,
    'rh500(%)' : 'rh500...' ,
    'ws500(m/s)' : 'ws500.m.s.' ,
    'wd500(deg)' : 'wd500.deg.',
    
})
ina_nwp_input_filtered
```

```python
temp_pred = model_xgb.predict(ina_nwp_input_filtered)
```

```python
temp_pred
```

```python
#OUTPUT

combined = pd.concat([ina_nwp_input['Date'], ina_nwp_input_filtered[['lokasi', 'suhu2m.degC.']], pd.Series(temp_pred, index = ina_nwp_input_filtered.index)], axis=1)
combined.columns = ['Date','lokasi', 'suhu2m.degC.', 'Prediction']
combined = combined.dropna()
combined.tail(15)
```

```python
import plotly.express as px
```

```python
px.line(combined[combined['lokasi'].isin([96001])],x='Date', y=['Prediction', 'suhu2m.degC.'], title='Temperature Observations over Time', color='lokasi', markers=True, orientation='v', line_shape='spline')
```
