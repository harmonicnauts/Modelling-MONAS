```python
new_df_unique = df_wmoid.merge(weather, on='lokasi')
new_df_unique
```

```python
new_df_unique_coord = new_df_unique[['lokasi','Nama UPT', 'LON', 'LAT']].drop_duplicates()
new_df_unique_coord
```

```python
m
```

```python
import geopandas
```

```python
new_df_unique_coord
```

```python
geometry = geopandas.points_from_xy(new_df_unique_coord.LON, new_df_unique_coord.LAT)
geo_df = geopandas.GeoDataFrame(
    new_df_unique_coord, geometry=geometry
)
geo_df = geo_df.reset_index(drop=True)
geo_df.head()
```

```python
geo_df['lokasi'][0]
```

```python
geo_df_list = [[point.xy[1][0], point.xy[0][0]] for point in geo_df.geometry]

i = 0
for coordinates in geo_df_list:
  m.add_child(
        folium.Marker(
            location=coordinates,
            popup="WMOID: "
            + str(geo_df['lokasi'][i])
            + "<br>"
            + "Nama UPT: "
            + str(geo_df['Nama UPT'][i])
            + "<br>"
            + "Coordinates: "
            + str(geo_df_list[i]),
          
        )
    )
  i = i + 1
```
